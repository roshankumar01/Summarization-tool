```python
from transformers import GPT2LMHeadModel , GPT2Tokenizer
```

```python
# load model
```

```python
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2" , pad_token_id =tokenizer.eos_token_id)
```

```python
# tokenize sentence
```

```python
sentence = "i love dogs"
inp_id = tokenizer.encode(sentence , return_tensors = 'pt')
```

```python
tokenizer.decode(inp_id[0])
```

```python
# generate text
```

```python
output = model.generate(inp_id , max_length = 50,num_beams= 2 , no_repeat_ngram_size =2,early_stopping=True)
```

```python
tokenizer.decode(output[0],skip_special_tokens=True)
```

```python
from transformers import pipeline
from bs4 import BeautifulSoup
import requests
```

```python
sumarizer  = pipeline('summarization')
```

```python
url = 'https://towardsdatascience.com/a-bayesian-take-on-model-regularization-9356116b6457'
r = requests.get(url)
```

```python
r.text
```

```python
soup = BeautifulSoup(r.text , 'html.parser')
results = soup.find_all(['h1','p'])
```

```python
text = [result.text for result in results]
article = ' '.join(text)
```

```python
# chunk
```

```python
article = article.replace(".",".<eos>")
article = article.replace("?","?<eos>")
article = article.replace("!","!<eos>")
sentence = article.split('<eos>')
```

```python
len(sentence)
```

```python
max_chunk = 500
current_chunk = 0
chunk = []

for sent in sentence:
    if len(chunk) == current_chunk +1:
        if len(chunk[current_chunk]) + len(sent.split(' ')) <= max_chunk:
            chunk[current_chunk].extend(sent.split(' '))
        else :
            current_chunk += 1
            chunk.append(sent.split(' '))
    else:
        chunk.append(sent.split(' '))
    
```

```python
len(chunk)
```

```python
chunk
```

```python
for chunk_id in range(len(chunk)):
    chunk[chunk_id] = ' '.join(chunk[chunk_id])
```

```python
chunk
```

```python
# sumarizing text
```

```python
res = sumarizer(chunk , max_length =100 , min_length = 20 ,do_sample=False)
```

```python
res
```

```python
summary = ' '.join(summ['summary_text'] for summ in res)
```
